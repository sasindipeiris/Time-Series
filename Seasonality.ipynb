Run this cell to set everything up!

```python
# Setup feedback system
from learntools.core import binder
binder.bind(globals())
from learntools.time_series.ex3 import *

# Setup notebook
from pathlib import Path
from learntools.time_series.style import *  # plot style settings
from learntools.time_series.utils import plot_periodogram, seasonal_plot

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess


comp_dir = Path('../input/store-sales-time-series-forecasting')

holidays_events = pd.read_csv(
    comp_dir / "holidays_events.csv",
    dtype={
        'type': 'category',
        'locale': 'category',
        'locale_name': 'category',
        'description': 'category',
        'transferred': 'bool',
    },
    parse_dates=['date'],
    infer_datetime_format=True,
)
holidays_events = holidays_events.set_index('date').to_period('D')

store_sales = pd.read_csv(
    comp_dir / 'train.csv',
    usecols=['store_nbr', 'family', 'date', 'sales'],
    dtype={
        'store_nbr': 'category',
        'family': 'category',
        'sales': 'float32',
    },
    parse_dates=['date'],
    infer_datetime_format=True,
)
store_sales['date'] = store_sales.date.dt.to_period('D')
store_sales = store_sales.set_index(['store_nbr', 'family', 'date']).sort_index()
average_sales = (
    store_sales
    .groupby('date').mean()
    .squeeze()
    .loc['2017']
)
```

# 2) Create seasonal features

Use `DeterministicProcess` and `CalendarFourier` to create:
- indicators for weekly seasons and
- Fourier features of order 4 for monthly seasons.

```python
y = average_sales.copy()


fourier = CalendarFourier(freq='M',order=4)
dp = DeterministicProcess(
    index=y.index,
    constant=True,
    order=1,
    seasonal=True,
    additional_terms=[fourier],
    drop=True,
)
X = dp.in_sample()

# Check your answer
q_2.check()
```

Now run this cell to fit the seasonal model.

```python
y = average_sales.copy()  # Target variable: average sales over time

# Create Fourier terms to model smooth monthly seasonality (using sine and cosine curves)
fourier = CalendarFourier(freq='M', order=4)

# Set up a deterministic process that includes:
# - A constant (intercept),
# - A linear time trend (order=1),
# - Seasonal dummies (e.g., for months),
# - And the Fourier seasonal terms created above
dp = DeterministicProcess(
    index=y.index,            # Time index from the sales data
    constant=True,            # Add an intercept term
    order=1,                  # Add a linear time trend (time, time² if order=2, etc.)
    seasonal=True,            # Add seasonal dummy variables (e.g., Jan, Feb, ...)
    additional_terms=[fourier],  # Add smooth Fourier seasonal curves
    drop=True                 # Drop terms to avoid multicollinearity
)

# Create the feature set (X) using the dates from y
X = dp.in_sample()
```
